```python
import requests
import re

url = "https://top.baidu.com/board"
```

上边是引用的库和url,下边是参数,这次我们方式选择get

```python
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64; rv:100.0) Gecko/20100101 Firefox/100.0",
}
params = {
    "tab": "realtime",
}
```

```python
response = requests.get(url=url, headers=headers, params=params)
print(response)
```

```python
print(response.text)
```

```python
obj_title = re.compile(r'<div class="c-single-text-ellipsis">(?P<title>.*?)</div>', re.S)
titles = obj_title.finditer(response.text)
```

预编译正则表达式

```python
lst = []
for title in titles:
    lst.append(title.group("title").replace(' ',''))
print(len(lst))
print(lst)
```

```python
response.close()
```
