数据爬取部分

```python
import json

import requests

url = "https://api.inews.qq.com/newsqa/v1/query/inner/publish/modules/list?modules=diseaseh5Shelf," \
      "chinaDayList,chinaDayAddList,nowConfirmStatis,provinceCompare"  # 这里的数据获取可以通过手动补充来实现
# 腾讯大好人,放开了查询接口
response = requests.post(url=url)
dict_all = json.loads(response.text)
print(type(dict_all))
```

```python
dict_all.keys()
```

```python
print(dict_all["data"].keys())
```

```python
print(dict_all['data']['chinaDayAddList'])
```

```python
dict_all['data']['diseaseh5Shelf'].keys()  # disease 疾病
```

```python
print(dict_all['data']['diseaseh5Shelf']['lastUpdateTime'])  # 上次数据更新时间
print(dict_all['data']['diseaseh5Shelf']['chinaTotal'])  # 当前汇总数据
print(dict_all['data']['diseaseh5Shelf']['chinaAdd'])  # 新增的数值
```

原来是打包获取数据,现在腾讯做成了接口,只需要提供目标参数,就能获取到对应数据
唯一的缺点就是获取路径太长了
